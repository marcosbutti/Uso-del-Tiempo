```python
import pandas as pd
import seaborn as sns
import numpy as np

data = pd.read_table("enut2021_diario.txt", sep="|")
data
```

```python
data.info()
```

```python
data.isnull().sum()  #Cantidad de Nulos por columna 
```

```python
#Reemplazo los nulos con 999 = no sabe/no contesta para manejar de forma numérica más simple
data = data.fillna(999)
data
```

```python
data["ACTIVIDAD_1"] = data["ACTIVIDAD_1"].astype("int64")  #Transformamos a entero los num flotantes
data.info()
```

```python
data["ACTIVIDAD_2"] = data["ACTIVIDAD_2"].astype("int64")
data["ACTIVIDAD_3"] = data["ACTIVIDAD_3"].astype("int64")
data.head()
```

```python
Act1 = data.groupby(data.ACTIVIDAD_1).count().sort_values("ID",ascending=False)
Act1.head(5)
#Como primer actividad las que más aparecen son: dormir, trabajar, comer y beber, 
#ver televisión, videos o transmisiones en vivo, reunirse con familiares o amigos...
```

```python
Act2 = data.groupby(data.ACTIVIDAD_2).count().sort_values("ID",ascending=False)
Act2.head(5)
#Como segunda actividad cambia cuales son las más comunes.
```

```python
Act3 = data.groupby(data.ACTIVIDAD_3).count().sort_values("ID",ascending=False)
Act3.head(5)
#Como tercer actividad también cambia pero predomina NoSabe/NoContesta
```

```python
#Cuanto duermen
dormir = data.loc[(data["ACTIVIDAD_1"] == 922)]
dormir
```

```python
tiempodormir = dormir.groupby(['ID', 'N_MIEMBRO']).size().reset_index(name='tiempo')
tiempodormir
```

```python
tiempodormir["tiempo"] = tiempodormir["tiempo"]*1/6  #Se transforman los bloques de 10 minutos en horas.
tiempodormir
```

```python
promdormir = tiempodormir["tiempo"].mean()
promdormir
```

```python
import matplotlib.pyplot as plt
```

```python
# Graficar los conteos
plt.figure(figsize=(18,8))
plt.scatter(tiempodormir["ID"], tiempodormir['tiempo'], marker=".")

# Trazar la línea con el promedio
plt.axhline(y=promdormir, color='red', linestyle='--')
plt.text(200000,promdormir + 0.5, " Promedio", color='red', fontsize=10)
plt.text(96000,promdormir,round(promdormir,2), color="red", fontsize=12)
plt.xlim(99000,plt.xlim()[1]) #Comienzo del eje X

# Configurar los títulos
plt.title("Tiempo de Dormir", fontsize=14, y=0.94)
plt.xlabel('ID')
plt.ylabel('Horas')

# Mostrar el gráfico
plt.show()
```
